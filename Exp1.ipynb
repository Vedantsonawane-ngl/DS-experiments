```python
import pandas as pd

# ===============================
# LOAD DATASET
# ===============================

df = pd.read_csv("sales_data_sample.csv", encoding="latin1")

print("First 5 rows:")
display(df.head())

# ===============================
# BASIC INSPECTION
# ===============================

print("\nShape:", df.shape)

print("\nLast 5 rows:")
display(df.tail())

print("\nInfo:")
df.info()

print("\nDescribe:")
display(df.describe())

# ===============================
# CONVERT ORDERDATE TO DATETIME
# ===============================

df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"])

print("\nAfter datetime conversion:")
df.info()

# ===============================
# MISSING VALUES
# ===============================

print("\nMissing values:")
display(df.isnull().sum())

# Fill missing STATE values
df["STATE"].fillna("Unknown", inplace=True)

# ===============================
# SALES OPERATIONS
# ===============================

print("\nTotal Sales:", df["SALES"].sum())
print("Average Sales:", df["SALES"].mean())
print("Max Sales:", df["SALES"].max())
print("Min Sales:", df["SALES"].min())

# ===============================
# FILTERING DATA
# ===============================

# 1. Orders from USA
usa_orders = df[df["COUNTRY"] == "USA"]

# 2. Orders in 2003
orders_2003 = df[df["YEAR_ID"] == 2003]

# 3. High Sales (> 5000)
high_sales = df[df["SALES"] > 5000]

# 4. USA orders in 2003
usa_2003 = df[(df["COUNTRY"] == "USA") & (df["YEAR_ID"] == 2003)]

# 5. Using isin()
selected_countries = df[df["COUNTRY"].isin(["USA", "France"])]

print("\nUSA Orders:")
display(usa_orders.head())

print("\n2003 Orders:")
display(orders_2003.head())

print("\nHigh Sales:")
display(high_sales.head())

print("\nUSA Orders in 2003:")
display(usa_2003.head())

print("\nUSA & France Orders:")
display(selected_countries.head())

# ===============================
# SORTING
# ===============================

sorted_sales = df.sort_values(by="SALES", ascending=False)
sorted_quantity = df.sort_values(by="QUANTITYORDERED")

# ===============================
# GROUPBY ANALYSIS
# ===============================

sales_by_country = df.groupby("COUNTRY")["SALES"].sum()
sales_by_dealsize = df.groupby("DEALSIZE")["SALES"].sum()
avg_sales_country = df.groupby("COUNTRY")["SALES"].mean()
multi_agg = df.groupby("COUNTRY")["SALES"].agg(["sum", "mean", "max"])

print("\nTotal Sales by Country:")
display(sales_by_country)

print("\nTotal Sales by Deal Size:")
display(sales_by_dealsize)

print("\nAverage Sales by Country:")
display(avg_sales_country)

print("\nMultiple Aggregations:")
display(multi_agg)

print("\n✅ DONE")
```
