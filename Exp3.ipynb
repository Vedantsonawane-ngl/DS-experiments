```python
import pandas as pd
import numpy as np
from scipy import stats

# load dataset
df = pd.read_csv("heart.csv")

# calculate z-score
z = np.abs(stats.zscore(df['chol']))

# find outliers
outliers_z = df[z > 3]

print("Outliers using Z-score:", len(outliers_z))
```

```python
Q1 = df['chol'].quantile(0.25)
Q3 = df['chol'].quantile(0.75)

IQR = Q3-Q1

lower = Q1-1.5*IQR
upper = Q3-7.5*IQR

outliners_iqr = df[(df['chol']<lower)|(df['chol']>upper)]
print("value =",len(outliners_iqr))
```
