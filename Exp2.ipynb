```python
import pandas as pd
import os
import csv

# 1. Create CSV file
def create_csv(filename="student_records.csv"):
    fields = ['Name', 'Branch', 'Year', 'CGPA']

    rows = [
        ['Nikhil', 'COE', '2', '9.0'],
        ['Sanchit', 'COE', '2', '9.1'],
        ['Aditya', 'IT', '2', '9.3'],
        ['Sagar', 'SE', '1', '9.5'],
        ['Prateek', 'MCE', '3', '7.8'],
        ['Sahil', 'EP', '2', '9.1']
    ]

    with open(filename, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(fields)
        writer.writerows(rows)

    print("CSV file created successfully!")


# 2. CSV → JSON
def csv_to_json(csv_file="student_records.csv", json_file="students.json"):
    df = pd.read_csv(csv_file)
    df.to_json(json_file, orient="records", indent=4)
    print("CSV converted to JSON successfully!")


# 3. CSV → Excel
def csv_to_excel(csv_file="student_records.csv", excel_file="students.xlsx"):
    df = pd.read_csv(csv_file)
    df.to_excel(excel_file, index=False)
    print("CSV converted to Excel successfully!")


# 4. Get file extension
def get_file_extension(file_name):
    ext = os.path.splitext(file_name)[1]
    print("File Extension:", ext)
    return ext


# 5. Display JSON content
def print_json(csv_file="student_records.csv"):
    df = pd.read_csv(csv_file)
    print(df.to_json(orient="records", indent=4))
```
